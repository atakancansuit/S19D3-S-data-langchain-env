```python
from dotenv import load_dotenv
dot_env_loaded = load_dotenv()
dot_env_loaded
```

```python
import os

gemini_api_key = os.environ.get("GEMINI_API_KEY", None)
google_cloud_project = os.environ.get("GOOGLE_CLOUD_PROJECT", None)

if not (gemini_api_key or google_cloud_project):
    print("Neither Gemini API key nor Google Cloud Project ID found in environment variables.")
    print("Please ask a TA for help.")
else:
    print("All good for now. Proceed to the next cell.")
```

```python
from google import genai

client = genai.Client()

client_created = True if client else False
client_created
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="Are you an LLM?",
)
chat_works = True if response else False
chat_works
```

```python
all_good = client_created and chat_works

from nbresult import ChallengeResult

result = ChallengeResult("setup", all_good=all_good)
result.write()
```

```python
print(result.check())
```
